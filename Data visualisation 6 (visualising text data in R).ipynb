As you can see, text data can be very messy and often requires a lot of cleaning before you can start to run any analysis. Another common method to remove unnecessary words is to use either <code>stemming</code> or <code>lemmatisation</code>. Stemming reduces words to their 'stem' which prefixes and suffixes add to in order to make variations of the root. For example 'wait' is the stem of 'waited', 'waiting', 'waits' and so on. Lemmatisation on the other hand groups together the inflected forms of a word. Where stemming chops off the extra parts of the word without knowledge of the word, Lemmatisation understands it's meaning before reducing to its lemma. For example this would know things like 'better' is derived from 'good'. Reducing words down to their stem or lemma also reduces the quantity of words.
