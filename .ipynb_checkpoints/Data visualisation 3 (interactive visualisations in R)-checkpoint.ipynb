Aside from the options available using <code>plotly</code>. The Jupyter notebooks also include a set of interactive widgets that can be used at add UI components to plots. 

```r
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
from matplotlib import pyplot as plt
```

Here is a simple example. We create a function called <code>get_value</code> that takes a parameter <code>x</code> and returns it.

```r
def get_value(x):
    return x
```

We can then pass the function into the inbuilt <code>interact</code> function with a default value (in this case <code>5</code>). As you move the slider, you will notice the value changing.

```r
interact(get_value, x=5);
```

There are other options too, for example a drop down (combo) box. In this example it will allow users to select a profession from a list of professions and output the selected value.

```r
def select_profession(prof='Nurse'):
    return prof
```

```r
interact(select_profession, prof=['Nurse', 'Doctor', 'Physio', 'OT', 'Pharmacist']);
```

These interactive elements can be added to notebooks that analyse data allowing users to select and filter different parts of the data without having to output many static versions of a given plot. Let's take a look at a more substantive example using the diabetes data from earlier.

```r
diabetes_data.head(10)
```

We can output a simple scatter plot of the <code>Glucose</code> against the <code>BMI</code> (Body Mass Index):

```r
plt.scatter(diabetes_data['Glucose'], diabetes_data['BMI']);
```

Let's add a line with a slope called <code>m</code>. In this example we will set it to <code>1</code> with a thickness/line width (<code>lw</code>) of <code>3</code> and make it red. 

```r
def show_plot(m=1):
    plt.scatter(diabetes_data['Glucose'], diabetes_data['BMI'])
    plt.plot(diabetes_data['Glucose'], m*diabetes_data['Glucose'], lw=3, color='red')
```

```r
show_plot(1)
```

Now we can pass it into the <code>interact</code> function allowing the user to alter the values of <code>m</code> using a slider.

```r
interact(show_plot, m=(0.1, 5, 0.1));
```

We could further enhance this with additional options. In this case we can add a parameter to allow the user to turn the line feature on or off. We call this <code>show_line</code> and set it to <code>False</code> by default. 

```r
def show_plot_line(m=1, show_line=False):
    plt.scatter(diabetes_data['Glucose'], diabetes_data['BMI'])
    if show_line:
        plt.plot(diabetes_data['Glucose'], m*diabetes_data['Glucose'], lw=3, color='red')
```

You can see below that as well as the slider, we now have a checkbox that allows us to turn the line on and off. Try clicking in the checkbox below:

```r
widgets.interact(show_plot_line, m=(0.1, 5, 0.1), show_line=False);
```

Let's take another quick look with some different data. This data is about cars and can be found with the <code>seaborn</code> visualisation library.

```r
import seaborn as sns
```

We load the data and output the first 5 records so you can see that it looks like. 

```r
car_data = sns.load_dataset('mpg')
car_data.head(5)
```

Here we have a function that outputs a KDE plot (Kernel Density Estimation) which is a way of visualising the distribution of observations in a dataset. We also set limits on the x and y values. We pass in a variable (<code>bw</code>) to set the band width (a smoothing parameter). 

```r
def change_bandwidth(bw=0.1):
    sns.kdeplot(car_data.horsepower, lw=2, fill=True, bw_adjust=bw)
    plt.xlim(-30, 300)
    plt.ylim(0, 0.03)
```

Next we pass the function to <code>interact</code> as previously. Move the slider to see the effect of changing the band width parameter on the visualisation.

```r
interact(change_bandwidth, bw=(0.1, 10));
```

Finally we can achieve the same with a different notation. The <code>@interact</code> function decorator can be used instead of the <code>interact</code> function like so:

```r
@interact(bw=(0.1, 10))
def change_bandwidth(bw=0.1):
    sns.kdeplot(car_data.horsepower, lw=2, fill=True, bw_adjust=bw)
    plt.xlim(-30, 300)
    plt.ylim(0, 0.03)
```

Finally lets create some simple data for a custom function.

```r
x = np.linspace(0, 2*np.pi, 1000)
```

We can then create a function called <code>show_sine_wave</code> that allows us pass in a value for <code>time</code> and frequency <code>fq</code>.

```r
def show_sine_wave(time=0, fq=1):
    y = np.sin(fq * (x+time))
    plt.plot(x, y)
```

```r
show_sine_wave(time=2, fq=2)
```

<div class="alert alert-block alert-info">
<b>Task 6:</b>
<br> 
    Use the <code>interact</code> feature to add sliders to the function to allow us to alter the time and frequency. Set a range for time between (0 and 20) and for frequency (0 to 30).
</div>

```r
@interact(time=(0, 20, 1), fq=(0, 30, 1))
def show_sine_wave(time=0, fq=1):
    y = np.sin(fq * (x+time))
    plt.plot(x, y)
```

There are many libraries to help you to add interactive feature to plots, from simple hover over to see information to 3D rotating plots and combining plots with UI elements for filtering. These sorts of plots are good for data exploration and demonstration of findings.
